# Imports and Utility Functions

```python
PROJECT_ROOT = '/home/aleksandar/projects/NLP_2021'
```

```python
import os

import pandas as pd
from gensim.models import Word2Vec
import nltk
nltk.download('punkt')
from nltk import word_tokenize

import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
def tokenize(data):
    data['tweet_tokens'] = data['tweet'].apply(lambda x: word_tokenize(x.lower()))
```

```python
def save(name, words, vectors):
    with open(f'{PROJECT_ROOT}/Exercises/2/data/{name}.txt', 'w+', encoding='utf-8') as doc:
        for word, vector in zip(words, vectors):
            doc.write(word + ' ' + ' '.join(str(value) for value in vector))
            doc.write('\n')
```

# Reading Data

```python
df = pd.read_csv(f"{PROJECT_ROOT}/Exercises/2/data/train.En.csv", usecols=['tweet', 'sarcastic']).dropna()
```

# Embeddings

```python
df.head()
```

```python
tokenize(df)
```

```python
sentences = df['tweet_tokens'].values
```

```python
model = Word2Vec(sentences, min_count=15, window=5, sg=1)
model.vector_size = 50
```

```python
vectors = model.wv.vectors
words = model.wv.index_to_key
```

```python
save('word2vecSG.iSarcasmEval.50d', words, vectors)
```

# Visualization

## Dimensionality Reduction

```python
tsne = TSNE(n_components=2, random_state=0)
vectors_2d = tsne.fit_transform(vectors)
```

```python
vectors_2d
```

```python
fig = px.scatter(x=vectors_2d[:,0], y=vectors_2d[:,1])
fig.show()
```

# Sentence Visualization

```python
model_sentence = Word2Vec(sentences, min_count=1, window=5, sg=0)
model_sentence.vector_size = 50
```

```python
df['embeddings'] = df['tweet_tokens'].apply(lambda x: model_sentence.wv[x])
df.head()
```

```python
df['avg_embeddings'] = df['embeddings'].apply(lambda x: x.mean(axis=0))
df.head()
```

## Dimensionality Reduction

```python
vectors = df['avg_embeddings'].values.tolist()
labels = df['sarcastic'].values.tolist()

tsne = TSNE(n_components=2, random_state=0)
vectors_2d = tsne.fit_transform(vectors)
```

```python
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], c=labels, cmap='jet', alpha=0.7)
plt.show()
```
